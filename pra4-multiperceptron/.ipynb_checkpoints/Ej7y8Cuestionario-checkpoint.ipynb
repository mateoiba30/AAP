```python
import h5py

# Abrir el archivo .h5
with h5py.File('mod_HOUSING.h5', 'r') as archivo:
    # Función recursiva para explorar la estructura y el contenido del archivo
    def imprimir_contenido(grupo, indent=0):
        for key in grupo.keys():
            item = grupo[key]
            print("  " * indent + key, ":", type(item))
            if isinstance(item, h5py.Dataset):  # Si es un dataset, imprime sus datos
                print("  " * (indent + 1) + "Datos:", item[()])
            elif isinstance(item, h5py.Group):  # Si es un grupo, explora recursivamente
                imprimir_contenido(item, indent + 1)

    # Llamar a la función para imprimir el contenido
    imprimir_contenido(archivo)
```

```python
import tensorflow as tf

@tf.keras.utils.register_keras_serializable(package='Custom', name='mse')
class MeanSquaredError(tf.keras.losses.MeanSquaredError):
    def __init__(self, name="mean_squared_error", **kwargs):
        super().__init__(name=name, **kwargs)
```

```python
from keras.layers import LeakyReLU
from keras.losses import MeanSquaredError

# Cargar el modelo, incluyendo la función de activación personalizada
model = tf.keras.models.load_model(
    'mod_HOUSING.h5', 
    custom_objects={
        'LeakyReLU': LeakyReLU,
        'MeanSquaredError': MeanSquaredError,
        'mse': MeanSquaredError  # Asegúrate de agregar esto si estás usando el nombre 'mse'
    }
)
```
