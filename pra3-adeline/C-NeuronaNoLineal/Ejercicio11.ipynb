la función del error binario realentiza el proceso de entrenamiento y tiene resultados similares
la función sigmoideal similar tanh en cuanto a precisión. La que es más rápida de entrenar es la función tanh.
los valores de alpha de 0.1, 0.2 y 0.3 en estos casos funcionaron igual de bien, habría que analizar en cada problema particualrmente para saber cuál valor de este elegir


```python
entrenamiento = 0.7
alpha = 0.3
func_act = "tanh"
func_costo = "ECM"
#en costo poner 'ECM' o 'EC_binaria'

reps = 30
suma_precisiones_test = 0
suma_precisiones_train = 0
suma_iteraciones = 0

for i in range (reps):

    %matplotlib inline
    import numpy as np
    import pandas as pd
    import time
    #from matplotlib import pyplot as plt
    from matplotlib import pylab as plt
    from IPython import display
    from sklearn import model_selection,preprocessing
    from grafica import *
    from ClassNeuronaGral import NeuronaGradiente

    datos = pd.read_csv("../../Datos/Vinos.csv", sep=";")
    #datos.head()

    datos["Class"].unique()

    mapeo = {
        "Class": {1:1, 2:0, 3:0}
    }
    datos.replace(mapeo,inplace=True)
    #datos.sample(15)

    #datos.info()

    testeo = 1 - entrenamiento

    X = np.array(datos.iloc[:,1:])
    T = np.array(datos['Class'])


    X_train, X_test, T_train, T_test = model_selection.train_test_split(
            X, T, test_size=testeo)

    normalizador= preprocessing.StandardScaler() #media y desvío
    X_train = normalizador.fit_transform(X_train)
    X_test  = normalizador.transform(X_test)
    
    if (func_act == 'tanh'): #los datos de predicción tienen que tener valores negativos porque la función tangencial va de 0 a 1 y necesita valores negativos para poder ajustar mejor los pesos
        T_train = 2*T_train-1
        T_test  = 2*T_test-1

    #print(X_train)

    ppn = NeuronaGradiente(alpha=alpha, n_iter=400, cotaE=1e-03, FUN=func_act, COSTO=func_costo, random_state=None, draw=1, title=["X, T"])
    ppn.fit(X_train, T_train)
    #no grafica nada porque tenemos muchas variables, por lo cual nos estamos pasando de las 3 dimensiones

    iteraciones = len(ppn.errors_)
    print("cantidad de iteraciones: ", len(ppn.errors_))
    suma_iteraciones += iteraciones
##print(f"suma actual: {suma_iteraciones}")

    #del entrenamiento:

    suma_precisiones = 0

    Y = ppn.predict(X_train)
    #print("Y = ", Y)
    #print("T = ", T_train)
    aciertos = sum(Y == T_train)
    #print("aciertos = ", aciertos)
    precision = 100*aciertos/X_train.shape[0]
    #print("%% de aciertos = %.2f %%" % (precision))
    suma_precisiones_train += precision

    print(f"precision promedio de {precision}")

    # plt.plot(range(1, len(ppn.errors_) + 1), ppn.errors_, marker='o')
    # plt.xlabel('Iteraciones')
    # plt.ylabel(func_costo)
    #plt.show()

    #del testeo
    #no debo repetir esta celda únicamente, debo repetir todo el entrenamiento del códgio para tener una nueva precisión


    Y = ppn.predict(X_test)
    #print("Y = ", Y)
    #print("T = ", T_test)
    aciertos = sum(Y == T_test)
    #print("aciertos = ", aciertos)
    precision = 100*aciertos/X_test.shape[0]
    #print("%% de aciertos = %.2f %%" % (precision))
    suma_precisiones_test += precision

    print(f"precision promedio de {precision}")

print("--------------------------------")
print(f"iteraciones promedio: {suma_iteraciones/reps}")
print(f"precisión de entrenamiento promedio: {suma_precisiones_train/reps}")
print(f"precisión de testeo promedio: {suma_precisiones_test/reps}")
```
