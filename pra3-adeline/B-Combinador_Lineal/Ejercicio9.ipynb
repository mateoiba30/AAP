```python
%matplotlib inline
import numpy as np
import pandas as pd
import time
#from matplotlib import pyplot as plt
from matplotlib import pylab as plt
from IPython import display
from sklearn import model_selection,preprocessing, metrics

from grafica import *
from ClassNeuronaLineal import NeuronaLineal
```

```python
datos= pd.read_csv('../../Datos/automobile-simple.csv')
datos.sample(15)
```

```python
print(datos['price'].isna().sum())
```

```python
price_promedio = datos["price"].mean()
print(price_promedio)
datos["price"] = datos["price"].fillna(price_promedio)
print(datos['price'].isna().sum())
```

```python
datos.info() #hay nulos en la columna de num-of-doors y horsepower
```

```python
datos = datos.dropna()
datos.info()
```

```python
marcas = datos.make.unique()
dicc_marcas = {}
i=0
for marca in marcas:
    dicc_marcas[marca]=i
    i+=1
print(dicc_marcas)
```

```python
fuel = datos["fuel-type"].unique()
dicc_fuel= {}
i=0
for f in fuel:
    dicc_fuel[f]=i
    i+=1
print(dicc_fuel)
```

```python
doors = datos["num-of-doors"].unique()
dicc_doors = {}
i=0
for d in doors:
    dicc_doors[d]=i
    i+=1
print(dicc_doors)
```

```python
body = datos["body-style"].unique()
dicc_body = {}
i=0
for b in body:
    dicc_body[b]=i
    i+=1
print(dicc_body)
```

```python
mapeo = {
    "make":dicc_marcas,
    "fuel-type":dicc_fuel,
    "num-of-doors":dicc_doors,
    "body-style": dicc_body
}
datos.replace(mapeo,inplace=True)
datos.sample(15)
```

```python
T = np.array(datos['price'])
datos_x = datos.drop('price', axis=1)
X = np.array(datos_x)
X = X.reshape(-1,1)
T = T.reshape(-1,1)

normalizar = 1
if normalizar:
    normalizador= preprocessing.MinMaxScaler()
    #normalizador= preprocessing.StandardScaler() #media y desvío
    X = normalizador.fit_transform(X)
    T = normalizador.fit_transform(T)
    
print(X)
print("----------")
print(T)
```

```python
X = X.reshape(201, 11)#volvemos a organizarlo en 11 columnas, para poder manipular bien los pesos de cada columna

nl = NeuronaLineal(alpha=0.01, n_iter=30, cotaE=10e-06)
MAX_INTENTOS = 50
W = np.zeros((MAX_INTENTOS,X.shape[1]))#shape[0] es el numero de filas y shape[1] el de columnas
#W tiene los pesos de cada columna, para cada iteración que hicimos
for i in range(MAX_INTENTOS):
    res=nl.fit(X, T)
    W[i,:]= nl.w_
```

```python
W[:,:] #vemos los pesos asignados a cada columna de X en cada iteración que tuvimos
```

```python
plt.plot(range(1, len(nl.errors_) + 1), nl.errors_, marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('Cantidad de actualizaciones')
plt.show()
```

```python
print("W = ", nl.w_)#esto muestra los pesos finales
print("b = ", nl.b_)
```
