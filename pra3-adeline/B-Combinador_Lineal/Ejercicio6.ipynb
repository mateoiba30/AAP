```python
%matplotlib inline
import numpy as np
import pandas as pd
import time
#from matplotlib import pyplot as plt
from matplotlib import pylab as plt
from IPython import display

from grafica import *
from ClassNeuronaLineal import NeuronaLineal
```

```python
datos= pd.read_csv('../../Datos/CCPP.csv')
T = np.array(datos['PE'])
X = np.array(datos['AT'])
X = X.reshape(-1,1)
print(X)
print(T)
```

```python
nl = NeuronaLineal(alpha=0.0001, n_iter=50, cotaE=10e-06, draw=1, title=['X: AT', 'Y: PE'])
# --- utilice random_state=None para que los pesos se inicializacen en forma aleatoria --
nl.fit(X, T)
```

```python
print("W = ", nl.w_)
print("b = ", nl.b_)
```

```python
plt.plot(range(1, len(nl.errors_) + 1), nl.errors_, marker='o')
plt.xlabel('Iteraciones')
plt.ylabel('Cantidad de actualizaciones')
plt.show()
```
