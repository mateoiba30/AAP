```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pylab as plt
from sklearn import model_selection, preprocessing
from sklearn import model_selection,preprocessing
from ClassNeuronaLineal import NeuronaLineal

# Datos de entrada
data = {
    'Horas de Entrenamiento': [2, 3, 4, 4, 5, 6, 6, 7, 7, 8, 10, 10],
    'Tiempo Promedio (min)': [30.5, 29.3, 26.7, 26.0, 25.5, 24.7, 23.5, 23.0, 22.5, 21.0, 21.0, 20.5]
}

df = pd.DataFrame(data)

print(df.corr())
```

```python
sns.regplot(x='Horas de Entrenamiento', y='Tiempo Promedio (min)', data=df, line_kws={"color": "red"})
plt.show()

# from google.colab import drive
# drive.mount('/content/drive')

# FUENTES_DIR = '/content/drive/My Drive/Aprendizaje Automático Profundo (AAP)/Colab Notebooks/Fuentes'
# DATOS_DIR = '/content/drive/My Drive/Aprendizaje Automático Profundo (AAP)/Colab Notebooks/Datos'


# import sys

# # Añadir el directorio a sys.path
# sys.path.append('/content/drive/My Drive/Aprendizaje Automático Profundo (AAP)/Colab Notebooks/Fuentes')


#Conviero al dataframe a numpy
T = np.array(df['Tiempo Promedio (min)'])
X = np.array(df['Horas de Entrenamiento'])

#Normalizo
X=X.reshape(-1,1)
T=T.reshape(-1,1)

normalizador= preprocessing.StandardScaler() #media y desvío
X = normalizador.fit_transform(X)
T = normalizador.fit_transform(T)



#Entreno
nl = NeuronaLineal(alpha=0.1, n_iter=1000, cotaE=10e-06, draw=1, title=['X: Horas de Entrenamiento', 'Y: Min'])
# --- utilice random_state=None para que los pesos se inicializacen en forma aleatoria --
nl.fit(X, T)

#Saco los coeficientes
print("W = ", nl.w_)
print("b = ", nl.b_)



nl.predict(1.2)  #Ver
```

```python
w = -0.86075054
b = -0.00736292

sumaError = 0

# Recorremos el DataFrame usando `iterrows()` que da acceso a cada fila como una serie
for index, row in df.iterrows():
    hora = row["Horas de Entrenamiento"]
    tiempo = row["Tiempo Promedio (min)"]
    tiempo_predicho = w * hora + b
    sumaError += (tiempo - tiempo_predicho) ** 2

# Calculamos el promedio del error cuadrático
print(sumaError / len(data))
```
