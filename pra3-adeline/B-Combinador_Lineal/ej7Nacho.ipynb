```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pylab as plt
from sklearn import model_selection, preprocessing
from sklearn import metrics
from ClassNeuronaLineal import NeuronaLineal

#Importa DF
df = pd.read_csv("../../Datos/AUTOS.csv",encoding='latin1')
datos2 = df
#Punto a, agregarle la media a los faltantes

#print(df.isnull().sum())

#Reemplazo faltantws con la media
df["price"] = df["price"].fillna(df["price"].mean())
df["normalized-losses"] = df["normalized-losses"].fillna(df["normalized-losses"].mean())
df["num-of-doors"] = df["num-of-doors"].fillna('four')
df["bore"] = df["bore"].fillna(df["bore"].mean())
df["stroke"] = df["stroke"].fillna(df["stroke"].mean())
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].mean())
df["peak-rpm"] = df["peak-rpm"].fillna(df["peak-rpm"].mean())
df["price"] = df["price"].fillna(df["price"].mean())

# Codificar 'make'
marcas = df["make"].unique()
dicc_marcas = {marca: i for i, marca in enumerate(marcas)}
print("Diccionario de marcas:", dicc_marcas)

# Codificar 'fuel-type'
fuel = df["fuel-type"].unique()
dicc_fuel = {f: i for i, f in enumerate(fuel)}
print("Diccionario de fuel-type:", dicc_fuel)

# Codificar 'num-of-doors'
doors = df["num-of-doors"].unique()
dicc_doors = {d: i for i, d in enumerate(doors)}
print("Diccionario de num-of-doors:", dicc_doors)

# Codificar 'body-style'
body = df["body-style"].unique()
dicc_body = {b: i for i, b in enumerate(body)}
print("Diccionario de body-style:", dicc_body)

# Codificar 'aspiration'
aspiration = df["aspiration"].unique()
dicc_aspiration = {a: i for i, a in enumerate(aspiration)}
print("Diccionario de aspiration:", dicc_aspiration)

# Codificar 'drive-wheels'
drive = df["drive-wheels"].unique()
dicc_drive = {d: i for i, d in enumerate(drive)}
print("Diccionario de drive-wheels:", dicc_drive)

# Codificar 'wheel-base'
wheel = df["wheel-base"].unique()
dicc_wheel = {w: i for i, w in enumerate(wheel)}
print("Diccionario de wheel-base:", dicc_wheel)

# Codificar 'engine-location'
engine_location = df["engine-location"].unique()
dicc_engine_location = {e: i for i, e in enumerate(engine_location)}
print("Diccionario de engine-location:", dicc_engine_location)

# Codificar 'engine-type'
engine_type = df["engine-type"].unique()
dicc_engine_type = {e: i for i, e in enumerate(engine_type)}
print("Diccionario de engine-type:", dicc_engine_type)

# Codificar 'fuel-system'
fuel_system = df["fuel-system"].unique()
dicc_fuel_system = {f: i for i, f in enumerate(fuel_system)}
print("Diccionario de fuel-system:", dicc_fuel_system)

# Codificar 'num-of-cylinders'
num_of_cylinders = df["num-of-cylinders"].unique()
dicc_num_of_cylinders = {n: i for i, n in enumerate(num_of_cylinders)}
print("Diccionario de num-of-cylinders:", dicc_num_of_cylinders)

# Crear el diccionario de mapeo
mapeo = {
    "make": dicc_marcas,
    "fuel-type": dicc_fuel,
    "num-of-doors": dicc_doors,
    "body-style": dicc_body,
    "aspiration": dicc_aspiration,
    "drive-wheels": dicc_drive,
    "wheel-base": dicc_wheel,
    "engine-location": dicc_engine_location,
    "engine-type": dicc_engine_type,
    "fuel-system": dicc_fuel_system,
    "num-of-cylinders": dicc_num_of_cylinders
}

# Reemplazar valores en el DataFrame con los valores codificados
df.replace(mapeo, inplace=True)

T = np.array(df['price'])

df = df.drop(columns=['price'])
X = np.array(df) # vector fila

#Normalizo
X = X.reshape(-1,1)
T = T.reshape(-1,1)

normalizar = 1
if normalizar:
    normalizador= preprocessing.MinMaxScaler()
    #normalizador= preprocessing.StandardScaler() #media y desv√≠o
    X = normalizador.fit_transform(X)
    T = normalizador.fit_transform(T)

#Entreno
```

```python
X = X.reshape(205, 25) #Para evitar que explote el reshape

nl = NeuronaLineal(alpha=0.001, n_iter=100, cotaE=10e-04)
nl.fit(X, T)

#Miro
print("pesos finales")

for i in range(25):
    print(df.columns[i], "W = ", nl.w_[i])#esto muestra los pesos finales
print("b = ", nl.b_)
```
